## Question 1: What factors influence the price of Airbnb listings in NYC?

We'll explore how variables like room_type, neighbourhood_group_cleansed, and number_of_reviews impact listing prices.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("cleaned_airbnb_nyc.csv")

# Set plot style
sns.set(style="whitegrid")

# Boxplot: Room type vs Price
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="room_type", y="price")
plt.title("Price Distribution by Room Type")
plt.yscale("log")  # Log scale to better view price variation
plt.xlabel("Room Type")
plt.ylabel("Price (log scale)")
plt.tight_layout()
plt.show()
```

Insight: Entire home/apt listings tend to be more expensive than private and shared rooms. This suggests that listing type significantly affects pricing strategy.

---

## Question 2: Which neighborhoods have the highest and lowest average prices?

Calculate average prices per neighborhood group to identify the most and least expensive areas.


```python
# Group by neighborhood group and compute average price
avg_price_neighborhood = df.groupby("neighbourhood_group_cleansed")["price"].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_neighborhood.index, y=avg_price_neighborhood.values, palette="viridis")
plt.title("Average Airbnb Price by Neighborhood Group")
plt.ylabel("Average Price ($)")
plt.xlabel("Neighborhood Group")
plt.tight_layout()
plt.show()
```

Insight: Manhattan is the most expensive borough on average, while the Bronx has the lowest listing prices. This aligns with housing market trends in NYC.
